```python
# 加载数据
import pandas as pd
import numpy as np
data = pd.read_csv('task2_data.csv')
data.head()
```

```python
#  x y 赋值
x = data.drop(['y'],axis=1)
y = data.loc[:,'y']
x.head()
y.head()
print(x.shape,y.shape)
```

```python
# 创建朴素贝叶斯模型并训练
from sklearn.naive_bayes import  CategoricalNB
model = CategoricalNB()
model.fit(x.values,y)
```

```python
# 训练数据的概率预测
y_predict_prob = model.predict_proba(x)
print(y_predict_prob)
```

```python
# 训练数据的预测
y_predict = model.predict(x)
print(y_predict)
```

```python
# 计算模型准确率
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y,y_predict)
print(accuracy)
```

```python
# 测试样本预测
x_test = np.array([[2,1,1,1,1]])
y_test_predict_prob = model.predict_proba(x_test)
y_test_predict = model.predict(x_test)
print(y_test_predict_prob)
print(y_test_predict)
```

```python
# 5个测试样本预测
x_test = np.array([[2,1,1,1,1],[2,1,1,1,0],[2,1,1,0,0],[2,1,0,0,0],
                    [2,0,0,0,0]])
print(x_test)
y_test_predict_prob = model.predict_proba(x_test)
y_test_predict = model.predict(x_test)
print(y_test_predict_prob)
print(y_test_predict)
```

```python
# 确认存储数据
print(x_test)
print(y_test_predict_prob)
print(y_test_predict)
print(x_test.shape,y_test_predict_prob.shape,y_test_predict.shape)
```

```python
# 数据组合
test_data_result = np.concatenate((x_test,y_test_predict_prob,y_test_predict.reshape(5,1)),
                                  axis=1)
print(test_data_result)
```

```python
# 格式转换
test_data_result = pd.DataFrame(test_data_result)
test_data_result.head()
```

```python
# 列名称替换
test_data_result.columns= ['score','school','award','gender',
                           'english','p0','p1','p2','y_test_predict']
test_data_result.head()
```

```python
# 保存数据
test_data_result.to_csv('test_data_result.csv')
```
